```python
df['Pays_de_vente'].fillna(value='', inplace=True)
```

```python
list_produit_pays = df[~(df['Pays_de_vente'].str.contains('en:'))]['Pays_de_vente'].tolist()
len(list_produit_pays)
```

```python
dict_pays = dict()
for elem in list_produit_pays:
    list_elem = elem.split(',')
    for elem in list_elem:
        #print()
        if elem.startswith(' '):
            pays = elem[1:]
        else:
            pays = elem
        if pays not in dict_pays:
            dict_pays[pays] = 0
        dict_pays[pays] += 1
```

```python
dict_pays.pop('')
df_pays = pd.DataFrame(list(dict_pays.items()))
df_pays.columns = ['country', 'count']
```

```python
# 6. Distribution des produits par nombre de pays de vente
plt.figure(figsize=(15,30))
plt.barh(df_pays['country'], df_pays['count']) 
plt.show()
```

```python
# Top pays hors France
top_pays_hors_france = df_pays.sort_values(by=['count'], ascending=False).head(20)[1:]
plt.figure(figsize=(12,7))
plt.barh(top_pays_hors_france['country'], top_pays_hors_france['count']) 
plt.show()
```

```python
new_df_vegetarien['Labels_certifications_recompenses'].fillna(value='', inplace=True)
```

```python
Labels_certifications_recompenses = new_df_vegetarien['Labels_certifications_recompenses'].tolist()
```

```python
Labels_certifications_recompenses_dict = dict()
for elem in Labels_certifications_recompenses:
    list_elem = elem.split(',')
    for elem in list_elem:
        #print()
        if elem.startswith(' '):
            label = elem[1:]
        else:
            label = elem
        if label not in Labels_certifications_recompenses_dict:
            Labels_certifications_recompenses_dict[label] = 0
        Labels_certifications_recompenses_dict[label] += 1
Labels_certifications_recompenses_dict.pop('')
```

```python
Labels_certifications_recompenses_df = pd.DataFrame(list(Labels_certifications_recompenses_dict.items()))
Labels_certifications_recompenses_df.columns = ['label', 'count']
plt.figure(figsize=(15,50))
plt.barh(Labels_certifications_recompenses_df['label'], Labels_certifications_recompenses_df['count']) 
plt.show()
```

```python
# Top labels
top_Labels_certifications_recompenses = Labels_certifications_recompenses_df.sort_values(by=['count'], ascending=False).head(20)
plt.figure(figsize=(12,7))
plt.barh(top_Labels_certifications_recompenses['label'], top_Labels_certifications_recompenses['count']) 
plt.show()
```
